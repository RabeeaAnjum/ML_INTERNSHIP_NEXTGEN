```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = pd.read_csv("store_customers.csv")
```

```python
data.head()
```

```python
data = data.dropna()
```

```python
data = data.drop_duplicates()
```

```python
data = data.drop("CustomerID", axis=1)
```

```python
scaler = StandardScaler()
data[data.select_dtypes(include='number').columns] = scaler.fit_transform(
    data.select_dtypes(include='number'))
```

```python
data.head()
```

```python
data.describe()
```

```python
data.mean(numeric_only=True)
```

```python
data.median(numeric_only=True)
```

```python
data.mode(numeric_only=True)
```

```python
data.var(numeric_only=True)
```

```python
data.hist(figsize=(10,8))
plt.show()
```

```python
data.boxplot(figsize=(10,6))
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()
```

```python
data["Gender"] = data["Gender"].map({"Male": 1, "Female": 0})
```

```python
data["Income_Age_Ratio"] = data["Annual Income (k$)"] / data["Age"]
```

```python
data.head()
```

```python
data.to_csv("cleaned_customer_data.csv", index=False)
```
